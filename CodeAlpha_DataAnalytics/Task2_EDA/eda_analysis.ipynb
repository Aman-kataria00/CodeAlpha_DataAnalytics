```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../Task1_WebScraping/quotes_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Quote_Length"] = df["Quote"].apply(len)
df.head()
```

```python
df["Quote_Length"].describe()
```

```python
plt.figure()
sns.histplot(df["Quote_Length"], bins=10)
plt.title("Distribution of Quote Lengths")
plt.xlabel("Number of Characters")
plt.ylabel("Frequency")
plt.show()
```

```python
top_authors = df["Author"].value_counts().head(5)

plt.figure()
top_authors.plot(kind="bar")
plt.title("Top 5 Authors by Number of Quotes")
plt.xlabel("Author")
plt.ylabel("Number of Quotes")
plt.show()
```

```python
## Exploratory Data Analysis Insights

- The dataset contains quotes scraped from a public website along with author names.
- No missing values were found, indicating good data quality.
- Quote lengths vary significantly, showing a mix of short and long quotations.
- A small number of authors contribute multiple quotes, while most authors appear once.
- Visual analysis helps understand content distribution and author frequency.
```
