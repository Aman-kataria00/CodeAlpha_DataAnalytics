```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Task2_EDA/cleaned_quotes_dataset.csv")
df.head()
```

```python
def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
df["Sentiment"] = df["Quote"].apply(get_sentiment)
df.head()
```

```python
sentiment_counts = df["Sentiment"].value_counts()

plt.figure()
sentiment_counts.plot(kind="bar")
plt.title("Sentiment Distribution of Quotes")
plt.xlabel("Sentiment")
plt.ylabel("Number of Quotes")
plt.show()
```

```python
df.to_csv("sentiment_results.csv", index=False)
```

```python
## Sentiment Analysis Insights

- Text sentiment was analyzed using TextBlob polarity scores.
- Quotes were classified into positive, negative, and neutral categories.
- Most quotes exhibit neutral to positive sentiment.
- Sentiment analysis helps understand emotional tone in textual data.
- This analysis can be extended to reviews, social media posts, or customer feedback.
```
